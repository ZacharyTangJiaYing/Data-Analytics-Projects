```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from selenium import webdriver
from selenium.common import exceptions
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementClickInterceptedException
from selenium.common.exceptions import ElementNotInteractableException
from selenium.common.exceptions import StaleElementReferenceException
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from webdriver_manager.chrome import ChromeDriverManager

import time 
import datetime
```

```python
# Settings for chrome driver
options = Options()

# options.add_argument('--headless')
options.add_argument("--no-sandbox")
options.add_argument("--disable-infobars")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--disable-browser-side-navigation")
options.add_argument("--disable-gpu") 
options.add_argument('--incognito')
options.add_argument("start-maximized")
options.add_argument("enable-automation")
driver = webdriver.Chrome(ChromeDriverManager().install(), options=options)

# data directory
data_path = input()

# target url
url = 'https://www.propertyguru.com.sg/property-for-sale'

# start driver
driver.get(url)
```

```python
# Variables to scrape

prop_listname = []
prop_address = []
prop_price = []
prop_bedrooms = []
prop_toilets = []
prop_sqft = []
prop_psf = []
prop_distance_to_MRT = []
prop_type = []
prop_lease = []
prop_built = []
prop_url = []
prop_agent = []
prop_list_recency = []
prop_scrape_time = []

list_of_features = [prop_listname, prop_address, prop_price, prop_bedrooms, prop_toilets, prop_sqft, prop_psf, prop_distance_to_MRT,
                   prop_type, prop_lease, prop_built]
```

```python
# Scraping algorithm

current_page = int(driver.find_element_by_xpath("//ul[@class='pagination']//li[@class='active']/a").text)
pagination_buttons = len(driver.find_elements_by_xpath("//ul[@class='pagination']/li"))
total_pages = int(driver.find_element_by_xpath("//ul[@class='pagination']/li[" + str(pagination_buttons - 1) + "]/a").get_attribute('data-page'))

while current_page <= total_pages:
    
    time.sleep(2)
    list_cards = driver.find_elements_by_xpath("//div[@class='listing-widget-new  ']/div")
    card = 1
    
    try:
        popup = driver.find_element_by_xpath('//*[@id="multipage-takeover-modal"]/div/a').click()
    
    except (NoSuchElementException, StaleElementReferenceException, ElementNotInteractableException):
        pass
    
    while True:
    
        try: 
            driver.find_element_by_xpath('//*[@id="hcaptcha_submit"]')
        except (NoSuchElementException):
            break
    
    while card <= len(list_cards):
        
        try:
            try:
                popup = driver.find_element_by_xpath('//*[@id="multipage-takeover-modal"]/div/a').click()
    
            except (NoSuchElementException, StaleElementReferenceException, ElementNotInteractableException):
                pass
            try:
                name = driver.find_element_by_xpath("//div[@class='listing-widget-new  ']/div[" + str(card) + "]/div[1]/div[2]/div[1]/div[1]/h3/a").get_attribute('title')
                prop_listname.append(name)
            except (NoSuchElementException):
                prop_listname.append(np.nan)
                pass

            try:
                address = driver.find_element_by_xpath("//div[@class='listing-widget-new  ']/div[" + str(card) + "]/div[1]/div[2]/div[1]/div[1]/p/span").text
                prop_address.append(address)
            except (NoSuchElementException):
                prop_address.append(np.nan)
                pass

            try:
                price = driver.find_element_by_xpath("//div[@class='listing-widget-new  ']/div[" + str(card) + "]/div[1]/div[2]/ul[1]/li/span[2]").text
                prop_price.append(price)
            except (NoSuchElementException):
                prop_price.append(np.nan)
                pass

            try:
                bedroom = driver.find_element_by_xpath("//div[@class='listing-widget-new  ']/div[" + str(card) + "]/div[1]/div[2]/ul[2]/li[1]/span[1]").text
                prop_bedrooms.append(bedroom)
            except (NoSuchElementException):
                prop_bedrooms.append(np.nan)
                pass

            try:
                toilet = driver.find_element_by_xpath("//div[@class='listing-widget-new  ']/div[" + str(card) + "]/div[1]/div[2]/ul[2]/li[1]/span[2]").text
                prop_toilets.append(toilet)
            except (NoSuchElementException):
                prop_toilets.append(np.nan)
                pass

            try:
                sqft = driver.find_element_by_xpath("//div[@class='listing-widget-new  ']/div[" + str(card) + "]/div[1]/div[2]/ul[2]/li[2]").text
                prop_sqft.append(sqft)
            except (NoSuchElementException):
                prop_sqft.append(np.nan)
                pass

            try:
                psf = driver.find_element_by_xpath("//div[@class='listing-widget-new  ']/div[" + str(card) + "]/div[1]/div[2]/ul[2]/li[3]").text
                prop_psf.append(psf)
            except (NoSuchElementException):
                prop_psf.append(np.nan)
                pass

            try:
                distance = driver.find_element_by_xpath("//div[@class='listing-widget-new  ']/div[" + str(card) + "]/div[1]/div[2]/ul[3]/li/").text
                prop_distance_to_MRT.append(distance)
            except (NoSuchElementException):
                prop_distance_to_MRT.append(np.nan)
                pass

            try:
                types = driver.find_element_by_xpath("//div[@class='listing-widget-new  ']/div[" + str(card) + "]/div[1]/div[2]/ul[4]/li[1]/span").text
                prop_type.append(types)
            except (NoSuchElementException):
                prop_type.append(np.nan)
                pass

            try:
                lease = driver.find_element_by_xpath("//div[@class='listing-widget-new  ']/div[" + str(card) + "]/div[1]/div[2]/ul[4]/li[2]/span").text
                prop_lease.append(lease)
            except (NoSuchElementException):
                prop_lease.append(np.nan)
                pass

            try:
                built = driver.find_element_by_xpath("//div[@class='listing-widget-new  ']/div[" + str(card) + "]/div[1]/div[2]/ul[4]/li[3]/span").text
                prop_built.append(built)
            except (NoSuchElementException):
                prop_built.append(np.nan)
                pass
            
            try:
                agent = driver.find_element_by_xpath("//div[@class='listing-widget-new  ']/div[" + str(card) + "]/div[2]/div/div/div[1]/div[1]/div/span").text
                prop_agent.append(agent)
            except (NoSuchElementException):
                prop_agent.append(np.nan)
                pass
            
            try:
                recency = driver.find_element_by_xpath("//div[@class='listing-widget-new  ']/div[" + str(card) + "]/div[1]/div[2]/div[2]").text
                prop_list_recency.append(recency)
            except (NoSuchElementException):
                prop_list_recency.append(np.nan)
                pass
            
            try:
                url = driver.find_element_by_xpath("//div[@class='listing-widget-new  ']/div[" + str(card) + "]/div[1]/div[2]/div[1]/div[1]/h3/a").get_attribute('href')
                prop_url.append(url)
            except (NoSuchElementException):
                prop_url.append(np.nan)
                pass
            
            prop_scrape_time.append(datetime.datetime.today().strftime("%m/%d/%Y, %H:%M:%S"))

            card += 1

        except:
            pass
    
    time.sleep(2)
    clear_output(wait=True)
    driver.execute_script("window.scrollTo(0, 5000)")
    current_pagination_buttons = len(driver.find_elements_by_xpath("//ul[@class='pagination']/li"))
    driver.find_element_by_xpath("//ul[@class='pagination']/li[" + str(current_pagination_buttons) + "]/a").click()

    print("Page [" + str(current_page) + "/" + str(total_pages) + "] scraped, " + str(round((current_page*100/total_pages),4)) + "% done")
    current_page += 1
```

```python
# Save to dataframe

df = pd.DataFrame(list(zip(prop_listname, prop_address, prop_price, prop_bedrooms, prop_toilets, prop_sqft, prop_psf, 
                           prop_distance_to_MRT, prop_type, prop_lease, prop_built, prop_url, prop_agent, prop_list_recency, prop_scrape_time)), 
                  columns=['name', 'address', 'price', 'bedrooms', 'toilets', 'sqft', 'psf', 'distance_to_mrt', 'property_type', 
                           'property_lease', 'build_date', 'url', 'agent', 'list_recency', 'scrape_time'])
df.head()
```

```python
df.to_csv(data_path + '/raw_data.csv')
```
